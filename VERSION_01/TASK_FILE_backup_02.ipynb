### Задача 2.1

Трафик-менеджер обратился к вам с просьбой определить качественные и некачественные источники трафика (**CAMPAIGN**) с точки зрения выполнения KPI по конверсии в выдачу и стоимости активированной карты.На выходе должен быть csv или Excel-файл с указанием N самых качественных и M самых некачественных источников (по CAMPAIGN по цене активированной карты). Все качественные источники должны вписываться в KPI по конверсии. N и M необходимо определить самостоятельно, соблюдая требования к значимости рассчитанных показателей.


### Решение

###### Идея
Заметим, что банк интересуют не просто выданные, но активированные карты: *Банк стремится минимизировать стоимость активированной карты*, *(по CAMPAIGN по цене активированной карты)*. С другой стороны, банк обязан платить поставщикам за выданную карту или заявку. Таким образом, основной ... много выданных карт/заявок, среди которых низкий процент активированных карт.


Поскольку задача имеет характер constrained optimization, мы можем разделить все источники на качественные/некачественные по признаку выполнения условия на KPI. Внутри группы качественных источников мы можем сортировать значения по средней стоимости активированной карты - чем дешевле карта, тем качественнее источник. Аналогично, можем оценить некачественные источники. Однако, в группе некачественных источников возможно наличие следуюших подгрупп:

    - нейтральная группа: 
процент активированных карт равен нулю, затраты на источник равны нулю.
    
    - отрицательная группа: 
процент активированных карт равен нулю, затраты на источник больше нуля. Такую группу можно считать самой некачественной, поскольку банк потратил деньги, не получив ничего взамен (при более детальном анализе можно еще делать поправку на длительность - можно предположить что источник трафика работает недостаточно долго, и в ближайшем будущем появятся клиенты, которые покроют затраты на данный источник).

Далее сгруппируем все данные внутри каждого уникального ```CAMPAIGN```. Теперь мы можем посчитать ... Заодно для каждого ```CAMPAIGN``` соберем все ```CAMPAIGN_MEDIUM```, что потенциально может представлять интерес для трафик-менеджера.

````python
def compute_final_campaign_df(df, conversion_condition="activated", threshold=5):
    
    columns = list(df.columns)
    columns.remove("CAMPAIGN_MEDIUM")
    columns.remove("CAMPAIGN")
    
    dct = {i: sum for i in columns}
    dct["CAMPAIGN_MEDIUM"] = list
    
    df = df.groupby("CAMPAIGN").agg(dct).drop(columns=["price", 
                                                       "positions"]).sort_values("ID_CONTACT_count", 
                                                                             ascending=False)
    
    df["mediums_count"] = df.CAMPAIGN_MEDIUM.apply(len)
    
    
    # average price
    df["ANY_ACTIVATED"] = df.ACTIVATED.apply(lambda x: x > 0)

    df["average_price"] = df.total_sum / df.ACTIVATED    
    ind = df.ANY_ACTIVATED == False
    df.loc[ind, "average_price"] = df[ind].total_sum
    
    df["FREE_CAMPAIGN"] = df.average_price.apply(lambda x: x == 0)
    
    
    # KPI
    df["conversion_percent"] = df[conversion_condition] / df['ID_CONTACT_count'] * 100
    df["KPI_is_done"] = df["conversion_percent"].apply(lambda x: x >= threshold)
    assert df.isna().any().sum() == 0
    
    return df
````

````python
def separate_groups(df):
    info_df = df.groupby(["KPI_is_done", 
                          "ANY_ACTIVATED", 
                          "FREE_CAMPAIGN"]).count().sort_index(ascending=False)[["ID_CONTACT_count"]]
    
    best_df = df[(df.KPI_is_done == True) &
                 (df.ANY_ACTIVATED == True) &
                 (df.FREE_CAMPAIGN) == True].sort_values("average_price", ascending=True)
    
    worst_df = df[(df.KPI_is_done == False) &
                  (df.ANY_ACTIVATED == False) &
                  (df.FREE_CAMPAIGN) == False].sort_values("average_price", ascending=False)
    
    return info_df, best_df, worst_df
````
